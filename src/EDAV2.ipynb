# Exploración de los datos

En esta sección, realizaremos una exploración inicial de los datos para entender su estructura y características. Esto incluye la carga de los datos, la visualización de las primeras filas, la identificación de valores faltantes y la descripción estadística de las variables, esto para darnos una idea de cómo están distribuidos los datos y qué transformaciones podrían ser necesarias para el análisis posterior.

```python
# Importamos las librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración de visualización
plt.style.use('ggplot')
%matplotlib inline
pd.set_option('display.max_columns', None)  # Mostrar todas las columnas
```

```python
# Cargamos el dataset
data = pd.read_csv('final_dataset_with_odds.csv')

# Revisión rápida de los datos
print("Dimensión del dataset:", data.shape)
print("Primeras filas del dataset:")
display(data.head())

# Información general del dataset
print("\nInformación del dataset:")
print(data.info())

# Estadísticas descriptivas
print("\nEstadísticas descriptivas:")
print(data.describe())
```

Informacion general de los datos --> Explicar

```python
# Revisión de valores faltantes
null_percent = data.isnull().mean() * 100
null_percent = null_percent[null_percent > 0].sort_values(ascending=False)

print("Porcentaje de valores nulos por columna:")
print(null_percent)

# Visualización de valores nulos
plt.figure(figsize=(12, 6))
sns.heatmap(data.isnull(), cbar=False, cmap='viridis')
plt.title('Mapa de valores faltantes en el dataset')
plt.show()

# Datos faltantes por columna
missing_data = data.isnull().sum()
missing_data = missing_data[missing_data > 0]
# Si no hay datos faltantes, no se mostrará nada
if not missing_data.empty:
    print("\nDatos faltantes por columna:")
    print(missing_data)
```

No hay valores nulos. (Tambien se puede ver en la parte de arriba)

```python
# Revisión de los datos duplicados
duplicates = data.duplicated().sum()
print(f"\nNúmero de filas duplicadas: {duplicates}")
```

Revisemos cual es el significado de cada columna del dataset:

Date: Fecha del partido (formato YYYY-MM-DD).
HomeTeam: Equipo local.
AwayTeam: Equipo visitante.
FTHG: Goles marcados por el equipo local al final del partido.
FTAG: Goles marcados por el equipo visitante al final del partido.
FTR: Resultado final del partido (H: victoria local, NH: No gana local).
HTGS: Goles marcados por el equipo local al medio tiempo.
ATGS: Goles marcados por el equipo visitante al medio tiempo.
HTGC: Goles recibidos por el equipo local al medio tiempo.
ATGC: Goles recibidos por el equipo visitante al medio tiempo.
HTP: Puntos obtenidos por el equipo local al medio tiempo.
ATP: Puntos obtenidos por el equipo visitante al medio tiempo.
